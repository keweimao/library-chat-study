```python
#Returns the source code
#Input: URL string
#Output: HTTP response object 
def get_sourceCode(url):
  try:
    session = HTMLSession()
    response = session.get(url)
    return response
  except requests.exceptions.RequestException as e:
    print(e)
```

```python
#scrapes google using the provided urls
def scrapeGoogleLinks(query):
  query = urllib.parse.quote_plus(query)
  response = get_sourceCode("https://www.google.com/search?q=" + query)
  links = list(response.html.absolute_links)
  googleDomainLinks = ('https://www.google.',
                       'https://google.',
                       'https://webcache.googleusercontent.',
                       'http://webcache.googleusercontent',
                       'https://policies.google.',
                       'https://support.google.',
                       'https://maps.google.')
  endLinks = ('pto=aue')
  for link in links[:]:
    if link.startswith(googleDomainLinks) or link.endswith(endLinks):
      links.remove(link)
  return links
```
